#Password Validation in Python (with REGEX)



*   At least 8 characters
*   uppercase letters: A-Z
*   lowercase letters:a-z
*   numbers: 0-9
*   any of the special characters: @#$%^&+=

Letters/numbers/special characters are optional







```python
regex = '[A-Za-z0-9@#$%^&+=]{8,}'
```

```python
pwd = input("Enter a password")
```

```python
pwd
```

```python
if re.fullmatch(regex,pwd):
  print("Congratulations, there is a match")
else:
  print("Sorry, there is no match")
```

You need to write regex that willvalidate a password to make sure it meets #ALL the following criteria



*   At least six characters long
*   contains a lowercase letter
*   contains an uppercase letter
*   contains a number

Valid passwords will only be alphanumeric characters



```python
regex2 = '(?=.*[A-Z])(?=.*[a-z])(?=.*[0-9]{6,})'
```

```python
if re.fullmatch(regex2, pwd):
  print("Match, YAy!")
else:
  print("Sorry No match")
```

**Question Mark**

Question mark ? is the option regex meaning 0 or 1 occurance.

Question mark ? can also be used in combination with other special characters and means something else.

Asterik * means 0 or more occurance of the preceding element.

By default, thr engine returns "more occurance" (greedy). If we want to force the 0 occurance, we use? after the * (non-greedy)

```python
import re
text1 = 'abcdefghijk'
```

```python
re.findall('ab.*?',text1)
```

```python
text2 = 'peter piper picked a peck of pickled peppers'
```

```python
re.findall('p.*?e.*?r',text2)
```

```python
text3 = '''If you want $1 in the year 1801. you Would have $180087791.41'''
```

```python
re.findall('(\$[0-9]+(\.(0-9)*)?)',text3)
```
